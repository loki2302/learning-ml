```python
%matplotlib inline

import pandas as pd
import zipfile

pd.options.display.max_columns = 100

# https://insights.stackoverflow.com/survey
zf = zipfile.ZipFile("developer_survey_2017.zip")
data = pd.read_csv(zf.open('survey_results_public.csv'))
data.head()
```

```python
# Distribution of professional levels
data["Professional"].value_counts().plot(kind="bar")
```

```python

d = data[pd.notnull(data["DeveloperType"])]
developer_type_lists = d["DeveloperType"].str.split(";")
developer_type_lists = developer_type_lists.apply(lambda t: list(map(lambda f: f.strip(), t)))
developer_types = []
for developer_types_list in developer_type_lists:
    developer_types += developer_types_list
import seaborn as sns
sns.countplot(y=developer_types)
```
